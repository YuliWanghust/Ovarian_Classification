```python
import os
import nibabel as nib

def get_max_min_coordinates(nii_file):
    # Load the NIfTI file
    nii_img = nib.load(nii_file)
    header = nii_img.header

    # Get the image dimensions
    dims = header.get_data_shape()

    # Calculate the maximum and minimum coordinates
    max_x = dims[0] - 1
    max_y = dims[1] - 1
    max_z = dims[2] - 1

    min_x = 0
    min_y = 0
    min_z = 0

    return (max_x, max_y, max_z), (min_x, min_y, min_z)

def main():
    # Specify the directory containing the .nii files
    directory = r"E:\Yuli\Projects\Ovarian\overian_classification\overian_classification\dataset\T1"

    # List all files in the directory
    files = os.listdir(directory)

    # Filter files ending with "T1_label.nii"
    t1_label_files = [f for f in files if f.endswith("T1_label.nii")]

    if not t1_label_files:
        print("No files found.")
        return

    for file in t1_label_files:
        file_path = os.path.join(directory, file)
        max_coords, min_coords = get_max_min_coordinates(file_path)
        print(f"File: {file}")
        print(f"Max coordinates: {max_coords}")
        print(f"Min coordinates: {min_coords}")
        print()

if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import nibabel as nib

def get_max_min_coordinates(nii_file):
    # Load the NIfTI file
    nii_img = nib.load(nii_file)
    data = nii_img.get_fdata()

    # Find indices where the value is 1
    nonzero_indices = np.where(data == 1)

    if len(nonzero_indices[0]) == 0:
        return None, None

    # Get maximum and minimum coordinates
    max_x = np.max(nonzero_indices[0])
    min_x = np.min(nonzero_indices[0])

    max_y = np.max(nonzero_indices[1])
    min_y = np.min(nonzero_indices[1])

    max_z = np.max(nonzero_indices[2])
    min_z = np.min(nonzero_indices[2])

    return (max_x, max_y, max_z), (min_x, min_y, min_z)

def main():
    # Specify the directory containing the .nii files
    directory = r"E:\Yuli\Projects\Ovarian\overian_classification\overian_classification\dataset\T1"

    # List all files in the directory
    files = os.listdir(directory)

    # Filter files ending with "T1_label.nii"
    t1_label_files = [f for f in files if f.endswith("T1_label.nii")]

    if not t1_label_files:
        print("No files found.")
        return

    for file in t1_label_files:
        file_path = os.path.join(directory, file)
        max_coords, min_coords = get_max_min_coordinates(file_path)
        if max_coords is None:
            print(f"No area with value 1 found in file: {file}")
        else:
            print(f"File: {file}")
            print(f"Max coordinates: {max_coords}")
            print(f"Min coordinates: {min_coords}")
            print()

if __name__ == "__main__":
    main()
```
